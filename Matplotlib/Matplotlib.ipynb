# Matplotlib Documentation (My Version)

- data visualization library 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_data = np.random.random(1000) * 100
Y_data = np.random.random(1000) * 100
```

# Scatter Plots
- c -> marker color
- s -> marker size
- alpha -> transparency : 0 - 1

```python
plt.scatter(X_data,Y_data, c="red", marker="*", s=100, alpha = 0.4)
```

# Line Plots
- lw -> line width

```python
years = [2006+x for x in range(16)]
years
```

```python
weights = [80, 83, 84, 85, 86, 82, 81, 79, 83, 80, 82, 82, 83, 81, 80, 79]
```

```python
plt.title("LinePlot")
plt.plot(years,weights, c="red", lw=2, linestyle="--") 
plt.grid(visible=True)
```

# Bar Graph

```python
Language = ["Python", "C++", "Java", "JavaScript", "Ruby"]
Votes = [500,300,375,1000,40]
```

```python
plt.bar(x=Language,height=Votes, color="black", width=0.7, edgecolor = "red", linewidth=3)
plt.title("Language Popularity")
plt.xlabel("Programming Language")
plt.ylabel("Votes")
```

# Histogram
- It groups numbers into ranges (called bins) and tells us how many data points fall into each range.
- Use a histogram to see how data is distributed across a range

- numpy.random.normal() -> generate random numbers that follows normal distribution (normal distribution is a way to describe how a set of data points are spread out)

```python
x = np.random.normal(loc=20, scale=1.5, size=1000)
x
```

```python
plt.hist(x,align="mid", bins = 20) #cumulative=True can also be used
```

# Pie Chart

```python
languages = ["Python", "C++", "Java", "Flutter", "GoLang"]
votes = [35,21,27,15,12]
explodes = [0,0,0,0.05]
```

```python
plt.pie(x=votes,labels=languages, explode=explodes, autopct="%.2f%%", pctdistance=0.6, startangle=180)
```
