# Replace Function

```python
weather_df = pd.read_csv("weather_data_2.csv")
weather_df
```

```python
new_weather_df = weather_df.replace({"temperature":"-99999","windspeed":"-99999","event":"no event"},np.nan) # ([-99999,-88888],np.nan)
new_weather_df
```

```python
new_weather_df_1 = weather_df.replace({
    -99999 : np.nan,
    "no event" : "Sunny"
})
new_weather_df_1
```

```python
new_weather_df = weather_df.replace({
    "temperature" : "[A-Za-z]",
    "windspeed" : "[A-Za-z]"
},"",regex=True)
new_weather_df
```

```python
student_df = pd.DataFrame({
    'score': ['exceptional','average', 'good', 'poor', 'average', 'exceptional'],
    'student': ['rob', 'maya', 'parthiv', 'tom', 'julian', 'erica']
})
student_df
```

# Concating Dataframes

```python
india_weather_df = pd.DataFrame({
    "city" : ["Mumbai","Delhi","Banglore"],
    "temperature" : [32,45,30],
    "humidity" : [80,60,78]
})
india_weather_df
```

```python
usa_weather_df = pd.DataFrame({
    "city" : ["New York","Chicago","Ohio"],
    "temperature" : [21,14,35],
    "humidity" : [68,65,75]
})
usa_weather_df
```

```python
# concat_df = pd.concat([india_weather_df,usa_weather_df], ignore_index = True) # ignores previous indexes
concat_df = pd.concat([india_weather_df,usa_weather_df], keys=["India","USA"]) # associating keys to dataframes
concat_df
```

```python
concat_df.loc["India"]
```

```python
temp_df = pd.DataFrame({
    "city" : ["Mumbai","Delhi","Banglore"],
    "temperature" : [32,45,30],
}, index=[0,1,2])
temp_df
```

```python
humidity_df = pd.DataFrame({
    "city" : ["Delhi","Banglore","Mumbai"],
    "humidity" : [80,60,78]
},index=[1,2,0])
humidity_df
```

```python
new_df = pd.concat([temp_df,humidity_df], axis=1)
new_df
```

```python
event_s = pd.Series(["Rainy","Sunny","Cloudy"], name="event")
event_s
```

```python
df = pd.concat([temp_df,event_s], axis=1)
df
```

# Merge Dataframes

```python
df1 = pd.DataFrame({
    "city" : ["new york", "chicago", "orlando","baltimore"],
    "temperature" : [21,14,35,38]
})
df1
```

```python
df2 = pd.DataFrame({
    "city" : ["chicago" , "new york" , "san francisco"],
    "humidity" : [65,68,71]
})
df2
```

```python
df3 = pd.merge(df1,df2,on="city",how="outer", indicator = True)
df3
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3 = pd.merge(df1,df2,on="city",how="outer",suffixes=["_left","_right"])
df3
```
