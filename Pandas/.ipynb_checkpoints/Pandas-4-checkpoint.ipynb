# Pivot 
- transforms the dataframe
  
  df.pivot(index = "", columns = "")

```python
weather_df = pd.read_csv("weather.csv")
weather_df
```

```python
weather_df.pivot(index="date", columns="city")
```

```python
weather_df.pivot(index="date", columns="city", values="humidity") # values = " " -> uses only those values to populate the dataframe
```

# Pivot table
- to summarize and produce appropriate data using aggregrate function

```python
df = pd.read_csv("weather2.csv")
df
```

```python
df.pivot_table(index="city",columns="date") # default aggfunc = "mean"
```

```python
df.pivot_table(index="city",columns="date", aggfunc="sum", margins=True)
```

# Grouper
- pd.Grouper allows you to group data by specific criteria, like time periods or intervals

```python
df = pd.read_csv("weather3.csv",parse_dates=["date"])
df
```

```python
# checking data on each month end
df.pivot_table(index=pd.Grouper(freq="ME", key="date"), columns="city")
```

# Melt 
- transforms wide format dataframe to long format dataframe

  id_vars -> columns to keep fixed
  
  value_vars -> columns to unpivot
  
  var_name -> name of the new column that will hold the original columns value
  
  value_name -> name of the column that will contain the original df value

```python
data = {
    "day": ["Monday", "Tuesday", "Wednesday"],
    "temperature": [22, 24, 19],
    "humidity": [60, 65, 70]
}

df = pd.DataFrame(data)
df
```

```python
df1 = pd.melt(df,id_vars=["day"], var_name="Metric", value_name = "Value")
df1
```

```python
df = pd.read_csv("weather_melt.csv")
df
```

```python
df1 = pd.melt(df,id_vars=["day"])
df1
```

# Stack Unstack
- df.stack() -> turns rows into columns
- df.unstack() -> turns columns into rows

```python
df = pd.read_excel("stocks.xlsx",sheet_name="stock_price",header=[0,1], index_col = 0) # there is two level of header 
df
```

```python
df1 = df.stack(level=1)
df1
```

```python
df1.unstack()
```

```python
df = pd.read_excel("stocks_3_levels.xlsx",header=[0,1,2], index_col=0)
df
```

```python
df.stack(level=2)
```

# Crosstab
- a crosstab in pandas is a table that shows the relationship between two or more categories by counting how often each combination of categories happens.
- pd.crosstab(index=,columns=)

```python
df = pd.read_excel("survey.xls")
df
```

- pip install xlrd -> to use .xls file

```python
pd.crosstab(df["Nationality"],df["Handedness"],normalize="index")
```

```python
pd.crosstab(df["Sex"],[df["Handedness"],df["Nationality"]],margins=True)
```

```python
pd.crosstab([df["Sex"],df["Nationality"]],df["Handedness"],margins=True, normalize=True)
```

```python
# *****
pd.crosstab([df["Sex"]],[df["Handedness"]], values=df["Age"],aggfunc="mean") # or aggfunc=np.average
```
