# Part - 2

```python
import pandas as pd
```

```python
# reading from excel file

def convert_people(cell):
    if(cell == "n.a."):
        return "sam walton"
    return cell

def convert_eps_cell(cell):
    if(cell == "not available"):
        return None
    return cell

stock_df = pd.read_excel("stock_data.xlsx","Sheet1",converters={
    "people" : convert_people,
    "eps" : convert_eps_cell
})
stock_df
```

```python
stock_df.to_excel("new.xlsx",sheet_name = "Stocks",index=False,startrow=1,startcol = 2)
```

- Writing two dataframes in single excel file in different sheets

```python
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather =  pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
```

```python
with pd.ExcelWriter("stock_weather.xlsx") as writer:
    df_stocks.to_excel(writer,sheet_name = "Sheet1", index=False)
    df_weather.to_excel(writer, sheet_name = "Sheet2",index=False)
```

- parse_dates = ["..."]

  converts string to date

```python
weather_df = pd.read_csv("weather_data_1.csv", parse_dates=["day"])
weather_df.set_index("day",inplace=True)
weather_df
```

```python
weather_df.loc["2017-01-01"]
```

# Handling missing values

- fillna

```python
new_weather_df = weather_df.fillna(0)
new_weather_df
```

```python
# using dictionary instead of a singe value
new_weather_df1 = weather_df.fillna({
    "temperature" : 0,
    "windspeed" : 0,
    "event" : "no event"
})
new_weather_df1
```

```python
# same value as previous day using .ffill() -> There is also df.bfill()
# new_weather_df2 = weather_df.ffill(limit = 1) -> forwarding value to one cell below only
new_weather_df2 = weather_df.ffill()
new_weather_df2
```

```python
# using df.interpolate() to fill missing values in better way
new_weather_df = weather_df.interpolate(method = "time")
new_weather_df
```

- dropna()

```python
new_weather_df = weather_df.dropna(thresh = 2) # how="all"
# thresh = x -> a row need to have atleast x non NaN values
new_weather_df
```

- Filling missing dates

```python
dr = pd.date_range("2017-01-01","2017-01-11")
idx = pd.DatetimeIndex(dr)
weather_df = weather_df.reindex(idx)
weather_df.ffill()
```
