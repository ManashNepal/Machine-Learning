```python
import pandas as pd
import numpy as np
```

# Replace Function

```python
weather_df = pd.read_csv("weather_data_2.csv")
weather_df
```

```python
new_weather_df = weather_df.replace({
    "temperature":-99999,"windspeed":-99999,"event":"no event"
},np.nan) # ([-99999,-88888],np.nan)
new_weather_df
```

```python
new_weather_df_1 = weather_df.replace({
    -99999 : np.nan,
    "no event" : "Sunny"
})
new_weather_df_1
```

```python
new_weather_df = weather_df.replace({
    "temperature" : "[A-Za-z]",
    "windspeed" : "[A-Za-z]"
},"",regex=True)
new_weather_df
```

```python
student_df = pd.DataFrame({
    'score': ['exceptional','average', 'good', 'poor', 'average', 'exceptional'],
    'student': ['rob', 'maya', 'parthiv', 'tom', 'julian', 'erica']
})
student_df
```

```python
new_student_df = student_df.replace({
    "exceptional" : 4,
    "good" : 3,
    "average" : 2,
    "poor" : 1
})
new_student_df
```

```python
# or we can use list
new_student_df1 = student_df.replace(["poor","average","good","exceptional"],[1,2,3,4])
new_student_df1
```
