```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

# Single Variable Linear Regression

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel("Area (sq. ft.)")
plt.ylabel("Price (USD$)")
plt.scatter(x=df["area"],y=df["price"], marker="+", color="red")
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(X=df[["area"]],y=df["price"])
```

```python
reg.predict(pd.DataFrame([[3300]], columns=["area"]))
```

```python
reg.predict(pd.DataFrame([[5000]],columns=["area"]))
```

# y = mx + b
- m -> coefficient
- b -> intercept

```python
# Coefficient -> m
reg.coef_
```

```python
# Intercept -> b
reg.intercept_
```

```python
df1 = pd.read_csv("areas.csv")
df1
```

```python
p = reg.predict(df1)
```

```python
df1["prices"] = p
```

```python
df1
```

```python
df1.to_csv("predictions.csv", index=False)
```

```python
%matplotlib inline
plt.xlabel("Area (sq. ft.)")
plt.ylabel("Price (USD$)")
plt.scatter(x=df["area"],y=df["price"], marker="+", color="red")
plt.plot(df["area"],reg.predict(df[["area"]]))
plt.show()
```

# Excercise

```python
df2 = pd.read_csv("canada_per_capita_income.csv")
df2.head()
```

```python
reg1 = linear_model.LinearRegression()
reg1.fit(df2[["year"]],df2["per capita income (US$)"])
```

```python
reg1.predict(pd.DataFrame([[2020]], columns=["year"]))
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("price")
plt.scatter(df2["year"],df2["per capita income (US$)"], c="red", marker="+")
plt.plot(df2["year"],reg1.predict(df2[["year"]]), color="green")
plt.show()
```
