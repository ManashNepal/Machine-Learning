# K Nearest Neighbors (KNN)

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(data=iris.data, columns = iris.feature_names)
df.head()
```

```python
df["target"] = iris.target
```

```python
df.head()
```

```python
df[df["target"] == 1].head()
```

```python
df[df["target"] == 2].head()
```

```python
df["flower_name"] = df["target"].apply(lambda x: iris.target_names[x])
df[50:].head()
```

```python
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
```

```python
import matplotlib.pyplot as plt
```

```python
# Setosa vs Versicolor - Sepal Length
```

```python
plt.scatter(df0["sepal length (cm)"], df0["sepal width (cm)"], c="red", marker="+")
plt.scatter(df1["sepal length (cm)"], df1["sepal width (cm)"], c="blue", marker="+")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
```

```python
# Setosa vs Versicolor - Petal Length
```

```python
plt.scatter(df0["petal length (cm)"], df0["petal width (cm)"], c="green", marker="+")
plt.scatter(df1["petal length (cm)"], df1["petal width (cm)"], c="violet", marker="+")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(["target", "flower_name"], axis = 1), df["target"], test_size = 0.2, random_state = 1)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
len(X_train)
```

```python
len(X_test)
```

# Creating KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
```

```python
params = {
    "n_neighbors" : [10,100,400,600,700]
}
clf = GridSearchCV(KNeighborsClassifier(),params,cv=3)
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, knn.predict(X_test))
cm
```

```python
import seaborn as sns

sns.heatmap(cm, annot = True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data
```

```python
digits.feature_names
```

```python
digits.target
```

```python
digits.target_names
```

```python
df = pd.DataFrame(digits.data, columns = digits.feature_names)
df.head()
```

```python
df["target"] = digits.target
df.head()
```

```python
train_X, test_X, train_y, test_y = train_test_split(df.drop(["target"], axis = 1), df["target"], test_size = 0.2, random_state=42)
```

```python
len(train_X)
```

```python
len(test_X)
```

```python
classifier = GridSearchCV(KNeighborsClassifier(), {
    "n_neighbors" : [3,5,7,10,15,20]
}, cv = 5)
classifier.fit(train_X, train_y)
```

```python
classifier.best_params_
```

```python
classifier.best_score_
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(train_X, train_y)
```

```python
# Confusion Matrix

cm1 = confusion_matrix(test_y, model.predict(test_X))
cm1
```

```python
sns.heatmap(cm1, annot = True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
```
