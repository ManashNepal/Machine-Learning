```python
import pandas as pd
from sklearn.datasets import load_digits
```

# Random Forest

```python
digits = load_digits()
dir(digits)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.gray()
plt.matshow(digits.images[0])
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
digits.target
```

```python
df["target"] = digits.target
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(["target"], axis = 1), df["target"], test_size = 0.2, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators = 80)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
import seaborn as sns
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, model.predict(X_test))
```

```python
sns.heatmap(cm, annot = True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
dir(iris)
```

```python
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
iris_df["target"] = iris.target
iris_df.tail()
```

```python
iris_model = RandomForestClassifier(n_estimators = 100)
X_train, X_test, y_train, y_test = train_test_split(iris_df.drop(["target"], axis = 1), iris_df["target"], test_size = 0.2)
```

```python
iris_model.fit(X_train,y_train)
```

```python
iris_model.score(X_test,y_test)
```

```python
cm1 = confusion_matrix(y_test,iris_model.predict(X_test))
cm1
```

```python
sns.heatmap(cm1, annot = True)
```
