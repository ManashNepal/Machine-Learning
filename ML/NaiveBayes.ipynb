# Titanic Survival

```python
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python
df.columns
```

```python
df = df[["Pclass", "Sex", "Age", "Fare", "Survived"]]
df.head()
```

```python
target = df["Survived"]
inputs = df.drop(["Survived"], axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
inputs["Sex"] = le.fit_transform(inputs["Sex"])
inputs.head()
```

```python
inputs.info()
```

```python
inputs.isna().sum()
```

```python
mean_val = inputs["Age"].mean()
mean_val
```

```python
inputs.fillna(mean_val, inplace = True)
```

```python
inputs.info()
```

```python
inputs[:10]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2, random_state=42)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

# Spam Email Detection

```python
spam_df = pd.read_csv("spam.csv")
spam_df.head()
```

```python
spam_df.groupby("Category").describe()
```

```python
spam_df["Category"] = spam_df["Category"].apply(lambda x: 1 if x == "spam" else 0)
spam_df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(spam_df[["Message"]], spam_df["Category"], test_size = 0.25, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.squeeze())
X_train_count.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
scam_model = MultinomialNB()
scam_model.fit(X_train_count, y_train)
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = v.transform(emails)
scam_model.predict(emails_count)
```

```python
X_test_count = v.transform(X_test.squeeze())
X_test_count.toarray()
```

```python
scam_model.score(X_test_count, y_test)
```

```python
from sklearn.pipeline import Pipeline
```

```python
clf = Pipeline([
    ("vectorizer", CountVectorizer()),
    ("nb", MultinomialNB())
])
```

```python
clf.fit(X_train.squeeze(), y_train)
```

```python
clf.predict(emails)
```

```python
clf.score(X_test.squeeze(), y_test)
```
