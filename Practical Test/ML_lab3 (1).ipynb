## Q-learning

```python
from sklearn.cluster import KMeans

kmean = KMeans(n_clusters = 5, n_init = 10, random_state = 42)
states = kmean.fit_predict(X_scaled)
states
```

```python
import numpy as np

n_states = 5
n_actions = 2

# Initialize q-table with zeroes
q_table = np.zeros((n_states, n_actions))

# Q-learning parameter
learning_rate = 0.1
epsilon = 0.2
n_epochs = 1000

q_table
```

```python
import random

y_np = y.to_numpy()

for epoch in range(n_epochs):
  for i in range(len(X_scaled)):
    # Get current state and true label
    state = states[i]
    true_label = y_np[i]

    # Epsilon-greedy action selection
    if random.uniform(0,1) < 0.1:
      action = random.choice([0,1])
    else:
      action = np.argmax(q_table[state])

    # Reward
    if action == true_label:
      reward = 1
    else:
      reward = -1

    # Update q-table
    old_value = q_table[state,action]
    new_value = old_value + learning_rate * (reward-old_value)
    q_table[state,action] = new_value

print("Q-learning Complete")
print(f"Q-table:\n{q_table}")
```

```python
y_pred = []

for i in range(len(X_scaled)):
  state = states[i]
  prediction = np.argmax(q_table[state])
  y_pred.append(prediction)

y_pred
```
